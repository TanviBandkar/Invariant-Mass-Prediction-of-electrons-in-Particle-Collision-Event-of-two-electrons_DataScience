###### But in particle physics the invariant mass is a characteristic of TOTAL ENERGY and MOMENTUM of the SYSTEM (Here system of two electrons)

```python
#Add another column of total energy of system
data['E_system']=data['E1'] + data['E2']
data.head()
```

##### Lets see with heatmap which parameter is most likely to affect invariant mass of the system

```python
co_relation=data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(co_relation,annot=True,fmt='0.1f')
```

###### As expected, invariant mass is more closely related to Total Energy E_system and transverse momentum pt

```python
#See how M is correlated with other feature parameters
co_rel_M=data.corr()
co_rel_M['M'].sort_values(ascending=False)
```

###### Since the Run and Event columns are just to specify run and event number and has nothing to do with invariant mass M, we can drop those

```python
data.drop(columns=['Run','Event'],axis=1,inplace=True)
```

### Splitting dataset

###### Target variable : M

```python
x=data.drop(labels='M',axis=1)
y=data['M']
```

```python
x
```

```python
y
```

###### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data)
```

```python
scaled_data
```

##### Splitting into training and testing data in ratio 80:20

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Modeling Techniques

### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred_LR=LR.predict(x_test)
y_pred_LR
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r_squared=r2_score(y_test,y_pred_LR)
r_squared
```

```python
mse_LR=mean_squared_error(y_test,y_pred_LR)
mse_LR
```

```python
#Root Mean Squared Error
rmse_LR=np.sqrt(mse_LR)
rmse_LR
```

###### check for accuracy of model

```python
plt.figure(figsize=(10,8))
min,max=y_test.min(),y_test.max()
sns.lineplot(x=[min,max],y=[min,max])
sns.scatterplot(x=y_test,y=y_pred_LR,color='yellow')
```

### 2. Decision Tree

```python
y_pred_DR=DR.predict(x_test)
y_pred_DR
```

```python
y_test
```

###### check for accuracy of model

```python
plt.figure(figsize=(10,8))
sns.lineplot(x=[min,max],y=[min,max])
sns.scatterplot(x=y_test,y=y_pred_DR,color='yellow')
```
